```python
import pandas as pd
product_data = pd.read_csv(r'C:\Users\hmina\OneDrive\デスクトップ\tarantula\第2回\vegetable_list.csv')
product_data.head()
```

```python
from object import Product, Price_Change_Product

# 商品数分処理
for num in range(len(product_data)):
    # データの読み込み
    # TODO: zip等使って簡潔にしたい
    name = product_data['name'][num]
    price = product_data['price'][num]
    area = product_data['area'][num]
    price_change = product_data['price_change'][num]

    # 値上げも値下げもないものは紹介のみ
    if price_change == 0:
        product = Product(name, price, area)
        product.introduction()

    # 値上げまたは値下げがあるものはアナウンス付き
    else:
        product = Price_Change_Product(name, price, area, price_change)
        product.notice()

print("以上で，お知らせを終わります．")
```
